```python
# Task 1 — Explore the Menu (5 marks)

categories = set()

for item in menu:
    categories.add(menu[item]["category"])

# Print categories
for cat in categories:
    print(f"======= {cat:<8} =======")
    for item in menu:
        if menu[item]["category"] == cat:
            price = menu[item]["price"]
            available = menu[item]["available"]

            if available:
                status = "[Available]"
            else:
                status = "[Unavailable]"

            print(f"{item:<15} ₹{price:<7.2f} {status}")

#calulation part- Total Items
total_items = len(menu)
print("="*20)
print("\nTotal items in menu:", total_items)

#Total no of available items
available_items = 0
for item in menu:
    if menu[item]["available"]:
        available_items += 1
print("\nTotal available items:", available_items)

#Most expensive item
max_price = 0
max_item = ""

for item in menu:
    price = menu[item]["price"]

    if price > max_price:
        max_price = price
        max_item = item

print("\nMost expensive item:", max_item, "₹", max_price)

#Items less than 150
print("\nItems under ₹150:")

for item in menu:
    price = menu[item]["price"]

    if price < 150:
        print(f"{item:<18} ₹{price:^7.2f}")
```

```python
# TASK 2 ==> CART OPERATIONS
cart = []

item_name = "Paneer Tikka"
quantity = 2

if item_name in menu:
    if menu[item_name]["available"]:
        price = menu[item_name]["price"]

        found = False

        for entry in cart:
            if entry["item"] == item_name:
                entry["quantity"] += quantity
                found = True
                print("Quantity updated")
                break

        if not found:
            cart.append({
                "item": item_name,
                "quantity": quantity,
                "price": price
            })
            print("Item added to cart")

    else:
        print("Item is unavailable")

else:
    print("Item does NOT exist")

print(cart)
```

```python
#Task 3 — Inventory Tracker with Deep Copy (6 marks)
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}
import copy

inventory_backup = copy.deepcopy(inventory)

# Change original inventory
inventory["Paneer Tikka"]["stock"] = 999

print("\nModified inventory:")
print(inventory["Paneer Tikka"])

print("\nBackup inventory:")
print(inventory_backup["Paneer Tikka"])

# original inventory restored
inventory["Paneer Tikka"]["stock"] = 10
print("\nRestored inventory:")
print(inventory["Paneer Tikka"])

#Simulate order fulfilment: deduct the quantities from the final cart in Task 2
for entry in cart:

    item = entry["item"]
    qty_needed = entry["quantity"]

    stock_available = inventory[item]["stock"]

    if stock_available >= qty_needed:
        inventory[item]["stock"] -= qty_needed
        print("\n=====Oder fulfilment simulation======")
        print(f"{item}: Updated Stock : {inventory[item]["stock"]}")

    else:
        print("\n=====Oder fulfilment simulation======")
        print(f"{item}: Only {stock_available} available, adjusting")
        inventory[item]["stock"] = 0

#Reorder Alerts
print("\n===== Reorder Alerts =====")
alert_found = False

for item in inventory:

    stock = inventory[item]["stock"]
    reorder_level = inventory[item]["reorder_level"]

    if stock <= reorder_level:
        print(f"⚠ Reorder Alert: {item} — Only {stock} unit(s) left (reorder level: {reorder_level})")
        alert_found = True

if not alert_found:
    print("No alerts")

#final check-deep copy
print("\n===============")
print("\nCurrent inventory:", inventory["Paneer Tikka"])
print("Backup inventory:", inventory_backup["Paneer Tikka"])
```

```python
#Task 4 — Daily Sales Log Analysis (6 marks)

#Revenue per day

print("\n===== Daily Revenue =====")

for date in sales_log:

    total_revenue = 0

    for order in sales_log[date]:
        total_revenue += order["total"]

    print(f"{date:<10}: ₹{total_revenue:.2f}")

#Best selling day

print("\n===== Best Selling Day =====")

max_revenue = 0
best_day = ""

for date in sales_log:

    total_revenue = 0

    for order in sales_log[date]:
        total_revenue += order["total"]

    if total_revenue > max_revenue:
        max_revenue = total_revenue
        best_day = date

print("Best-selling day:", best_day, "with revenue ₹", max_revenue)


#Most ordered item

print("\n===== Most Ordered Item =====")

item_count = {}
for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item in item_count:
                item_count[item] += 1
            else:
                item_count[item] = 1

max_count = 0
most_ordered_item = ""

for item in item_count:

    if item_count[item] > max_count:
        max_count = item_count[item]
        most_ordered_item = item

print("Most ordered item:", most_ordered_item, "ordered", max_count, "times")


```
